# use 0 for not shipped when quantity 0

```python
pf['item_status']=pf['item_status'].replace(to_replace=np.nan,value=0)
pf['item_status']
```

# use currency INR

```python
pf['currency']=pf['currency'].replace(to_replace=np.nan,value='INR')
pf['currency']
```

# item price is 0 for all nan(BCZ NO QUANTITY)

```python
pf['item_price']=pf['item_price'].replace(to_replace=np.nan,value=0)
pf['item_price']
```

```python
pf.isnull().sum()
```

```python
pf['item_tax']=pf['item_price'].replace(to_replace=np.nan,value=0)
pf['item_tax']
```

```python
pf.isnull().sum()
```

# item tax is 450 when item price is 2950

```python
pf['item_tax']=pf['item_price'].replace(to_replace=2950,value=450)
pf['item_tax']
```

```python
pf.isnull().sum()
```

# ship price 50 for TN and 100 for other state

# 

```python
pf.loc[pf['ship_state'] == 'TAMIL NADU', 'shipping_price'] = 50
pf['shipping_price']
pf.loc[pf['ship_state'] != 'TAMIL NADU', 'shipping_price'] = 100
pf['shipping_price']
```

```python
pf.isnull().sum()
```

# use tax rate as 0.10(default)   (using ship_price & tax_rate calculate tax)

```python
tax_rate = 0.10
pf['shipping_tax']=pf['shipping_price']*tax_rate
pf['shipping_tax']
```
