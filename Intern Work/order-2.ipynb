```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load your cleaned dataset
pf = pd.read_csv('cleaned_datasets.csv')

# Ensure 'purchase_date' column is in datetime format
pf['purchase_date'] = pd.to_datetime(pf['purchase_date'], errors='coerce')

# Drop rows where 'purchase_date' couldn't be converted to datetime (if any)
pf = pf.dropna(subset=['purchase_date'])

# Group by 'purchase_date' and sum 'quantity'
grouped_data = pf.groupby(pf['purchase_date'].dt.date)['quantity'].sum().reset_index()

# Convert 'purchase_date' to datetime format if it's in date format
grouped_data['purchase_date'] = pd.to_datetime(grouped_data['purchase_date'])

# Feature engineering: Extract date-related features
grouped_data['day_of_week'] = grouped_data['purchase_date'].apply(lambda x: x.dayofweek)
grouped_data['day_of_month'] = grouped_data['purchase_date'].apply(lambda x: x.day)
grouped_data['month'] = grouped_data['purchase_date'].apply(lambda x: x.month)

# Prepare data for linear regression
X = grouped_data[['day_of_week', 'day_of_month', 'month']]  # Features
y = grouped_data['quantity']  # Target variable

# Fit linear regression model
lr_model = LinearRegression()
lr_model.fit(X, y)

# Predict quantities for all historical dates
grouped_data['predicted_quantity'] = lr_model.predict(X)

# Predict quantities for the next 5 days
# Assuming future dates are consecutive days after the last date in the dataset
future_dates = pd.date_range(start=grouped_data['purchase_date'].max() + pd.Timedelta(days=1), periods=5, freq='D')
future_data = pd.DataFrame({'purchase_date': future_dates})
future_data['day_of_week'] = future_data['purchase_date'].apply(lambda x: x.dayofweek)
future_data['day_of_month'] = future_data['purchase_date'].apply(lambda x: x.day)
future_data['month'] = future_data['purchase_date'].apply(lambda x: x.month)

# Predict quantities for future dates
future_data['predicted_quantity'] = lr_model.predict(future_data[['day_of_week', 'day_of_month', 'month']])

# Display quantities sold for all dates
print("Historical Quantities Sold:")
print(grouped_data[['purchase_date', 'quantity']])

# Display predicted quantities for the next 5 days
print("\nPredicted Quantities for Next 5 Days:")
print(future_data[['purchase_date', 'predicted_quantity']])

# Plotting historical and predicted quantities
plt.figure(figsize=(12, 6))

# Plot actual quantities
plt.plot(grouped_data['purchase_date'], grouped_data['quantity'], marker='o', linestyle='-', color='b', label='Actual')

# Plot predicted quantities for all dates (including historical and future)
plt.plot(grouped_data['purchase_date'], grouped_data['predicted_quantity'], marker='o', linestyle='-', color='r', label='Predicted (Historical)')

# Plot predicted quantities for future dates
plt.plot(future_data['purchase_date'], future_data['predicted_quantity'], marker='o', linestyle='-', color='g', label='Predicted (Next 5 Days)')

plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.title('Actual vs Predicted Quantity Sold (Linear Regression)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# time series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

# Load your cleaned dataset
pf = pd.read_csv('cleaned_datasets.csv')

# Ensure 'purchase_date' column is in datetime format
pf['purchase_date'] = pd.to_datetime(pf['purchase_date'], errors='coerce')

# Drop rows where 'purchase_date' couldn't be converted to datetime (if any)
pf = pf.dropna(subset=['purchase_date'])

# Group by 'purchase_date' and sum 'quantity'
grouped_data = pf.groupby(pf['purchase_date'].dt.date)['quantity'].sum().reset_index()

# Convert 'purchase_date' to datetime format if it's in date format
grouped_data['purchase_date'] = pd.to_datetime(grouped_data['purchase_date'])

# Feature engineering: Extract date-related features and lagged quantities
grouped_data['day_of_week'] = grouped_data['purchase_date'].dt.dayofweek
grouped_data['day_of_month'] = grouped_data['purchase_date'].dt.day
grouped_data['month'] = grouped_data['purchase_date'].dt.month

# Adding lagged quantities as features (autoregressive model)
num_lags = 7  # Using 7 days of lagged quantities as an example
for i in range(1, num_lags + 1):
    grouped_data[f'lag_{i}_quantity'] = grouped_data['quantity'].shift(i)

# Remove rows with NaN resulting from shift operation
grouped_data = grouped_data.dropna()

# Prepare data for linear regression
X = grouped_data[['day_of_week', 'day_of_month', 'month'] + [f'lag_{i}_quantity' for i in range(1, num_lags + 1)]]
y = grouped_data['quantity']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Initialize linear regression model
lr_model = LinearRegression()

# Fit linear regression model
lr_model.fit(X_train, y_train)

# Predict quantities for all historical dates
grouped_data['predicted_quantity'] = lr_model.predict(X)

# Predict quantities for the next 5 days
# Assuming future dates are consecutive days after the last date in the dataset
future_dates = pd.date_range(start=grouped_data['purchase_date'].max() + pd.Timedelta(days=1), periods=5, freq='D')
future_data = pd.DataFrame({'purchase_date': future_dates})
future_data['day_of_week'] = future_data['purchase_date'].dt.dayofweek
future_data['day_of_month'] = future_data['purchase_date'].dt.day
future_data['month'] = future_data['purchase_date'].dt.month

# Adding lagged quantities for future dates as features
for i in range(1, num_lags + 1):
    future_data[f'lag_{i}_quantity'] = grouped_data['quantity'].shift(i).iloc[-1]

# Predict quantities for future dates
future_data['predicted_quantity'] = lr_model.predict(future_data[['day_of_week', 'day_of_month', 'month'] + [f'lag_{i}_quantity' for i in range(1, num_lags + 1)]])

# Display quantities sold for all dates
print("Historical Quantities Sold:")
print(grouped_data[['purchase_date', 'quantity']])

# Display predicted quantities for the next 5 days
print("\nPredicted Quantities for Next 5 Days:")
print(future_data[['purchase_date', 'predicted_quantity']])

# Plotting historical and predicted quantities
plt.figure(figsize=(12, 6))

# Plot actual quantities
plt.plot(grouped_data['purchase_date'], grouped_data['quantity'], marker='o', linestyle='-', color='b', label='Actual')

# Plot predicted quantities for all dates (including historical and future)
plt.plot(grouped_data['purchase_date'], grouped_data['predicted_quantity'], marker='o', linestyle='-', color='r', label='Predicted (Historical)')

# Plot predicted quantities for future dates
plt.plot(future_data['purchase_date'], future_data['predicted_quantity'], marker='o', linestyle='-', color='g', label='Predicted (Next 5 Days)')

plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.title('Actual vs Predicted Quantity Sold (Autoregressive Linear Regression)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
